```python
from pathlib import Path
import sqlite3
```

```python
data_dir = Path.home() / 'Dropbox' / 'Python' / 'Sudoku' / 'data' / 'output'
```

```python
conn = sqlite3.connect(data_dir / 'SudokuPuzzles.db')
cursor = conn.cursor()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS UnsolvedPuzzles (
        Book TEXT NOT NULL,
        PuzzleID INTEGER
    )
""")
conn.commit()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS DoubleSudoku (
        Book TEXT NOT NULL,
        PuzzleID INTEGER,
        Status TEXT
    )
""")
conn.commit()
```

```python
conn.close()
```
